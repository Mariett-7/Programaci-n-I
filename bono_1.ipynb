Manejo de módulos integrados

NumPy: se usa principalmente para trabajar con datos númericos de forma rápida, eficiente y estructurada. Un objeto tipo NumPy es un arreglo base para hacer matemáticas y manipular datos en Phyton.

Pandas: se usa para trabajar con datos estructurales (Como tablas de Excel o bases de datos). Un objeto tipo pandas permite cargar, limpiar, transformar y analizar datos. Lo usamos para bases de datos relacionales.

La diferencia entre NumPy y Pandas radica en su propósito y nivel de uso: NumPy es una liberia enfocada en el cálculo numérico eficiente mediante arrays multidimensionales, ideal para operaciones matemáticas rápidas, álgebra lineal y procesamiento de datos puramente numéricos; mientras que Pandas es una libreria de más alto nivel diseñada para la manipulación y análisis de datos estructurados, como tablas tipo Excel o bases de datos, permitiendo trabajar con columnas con nombres, datos de distintos tipos, limpieza, filtrado y agrupación; en la práctica, Pandas se apoya internamente de NumPy, por lo que ambos se complementan y se utilizan juntos en la mayoria de flujos de ciencia de datos.

```python
data = {
    'nombre': ['Ana', 'Luis', 'Carlos', 'Ana', None],
        'edad': [23, np.nan, 35, 23, 40],
        'salario': [1000, 2000, None, 1000, 5000],
        'ciudad': ['Bogotá', 'Medellín', 'Cali', 'Bogotá', 'Cali']
    }
data
```

```python
data1 = pd.DataFrame(data) #Así tranformamos un diccionario en una tabla de Pandas
data1
```

```python
data1.isnull() #Valores Faltantes
#El médoto isnull, muestra true si el valor es faltante y false si el valor no es faltante
```

```python
data1.isnull().sum() #El método sum, suma los valores por columnas: 1 si es faltante, 0 si no.
```

```python
data2 = data1.dropna() #El modelo Dropna, elimina las filas con valores faltantes
data2
```

```python
data1.dropna(subset=['salario']) #Se eliminan las filas con valores faltantes, únicamente
#en la columna salario
```

```python
data1['edad'] = data1['edad'].fillna(data1['edad'].mean())
data1['nombre'] = data1['nombre'].fillna('Desconocido')
data1
#Se están imputando datos faltantes, es decir que ponga la media aritmética de la columna de edad de data1
#Con la columna nombre la esta reemplazando con la misma columna nombre, pero a los valores faltantes les pone desconocido
```

```python
data1['ciudad'] = data1['ciudad'].replace('Bogotá', 'BOGOTA')
data1
#El modelo replace o reemplazo, lo que hace es reemplazar la palabra Bogotá por BOGOTA en Mayúscula, en la columna ciudad
#de data1
```

```python
data4 = data1['ciudad'].drop_duplicates()
data4
#Se toma data1 sin duplicados
```

```python
dfn = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'nombre': ['Ana', 'LUIS', 'Carlos', 'ana'],
        'edad': [23, 30, 35, 28],
        'salario': [1000, 2000, 3000, 1500],
        'ciudad': ['Bogotá', 'Cali', 'Bogotá', 'Medellín']
    })
dfn
```

```python
dfn['salario_anual'] = dfn['salario'] * 12 #aplicar función
dfn['edad_doble'] = dfn['edad'].apply(lambda x: x * 2)
dfn
#Estamos agregando una nueva columna: una que es el salario anual y la otra que es el doble de la edad
```

```python
dfn = dfn.rename(columns= {'edad': 'age'}) #Renombrar columnas, entonces la columna edad la renombra como edad en inglés
dfn
```

```python
dfn['nombre'] = dfn['nombre'].str.upper() #Convierte al objeto en letras mayúsculas como son las cadenas del nombre
dfn['nombre'] = dfn['nombre'].str.strip() #Limpia las cadenas del nombre y eliminando los espacios en blanco
dfn['nombre'] = dfn['nombre'].str.replace('ANA', 'JOANA') #Replace reemplaza a ana por ANA
```

```python
dfn
```

```python
grupo = dfn.groupby('ciudad')['salario'].mean()
grupo
#Este método toma la columna ciudad y todos los elementos de la columna salario, para calcular la media aritmética del salario
#Agrupa todos los elementos de ciudad y calcula a esos grupos la media aritmética
```

```python
ordenado = dfn.sort_values(by= 'age')
ordenado
#Ordenar los datos de la base de datos
```

```python
dfn1 = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'departamento': ['Ventas', 'IT', 'RRHH', 'Marketing']
    })

merge = pd.merge(dfn, dfn1, on= 'id')
merge
```
